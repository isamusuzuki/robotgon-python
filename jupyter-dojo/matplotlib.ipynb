## サンプルコード1号


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y)
```

## いろいろな図を作成する

```python
# sin関数を2次元平面に描画する
x = np.arange(-3, 3, 0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
# 散布図
x = np.random.randn(30)
y = np.sin(x) + np.random.randn(30)
plt.plot(x, y, 'ro')
# 'o' ... 小さい円
# 'ro' ... 赤い円
```

```python
# 正規分布に従う乱数をヒストグラムで描画
plt.hist(np.random.randn(100))
```

```python
# 2次元の巨大な100x100行列を画像として可視化
plt.imshow(np.random.randn(100, 100))
plt.title("100x100 matrix")
```

```python
# y軸の表示幅を変更
x = np.arange(0, 10, 0.1)
y = np.exp(x)
plt.plot(x, y)
plt.title("exponential function: $ y = e^x $")
plt.ylim(0, 5000)
```
