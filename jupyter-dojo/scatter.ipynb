# 散布図を描画する


```python
# 最も単純な散布図の例
import matplotlib.pyplot as plt
import numpy as np

# データを生成する
x = np.random.rand(100)
y = np.random.rand(100)

# 描画領域を作成する
# ひとつの描画領域には複数の座標軸を作成できる
# axes = axisの複数形
figure = plt.figure()

# 座標軸をひとつ作成する
axis = figure.add_subplot(1,1,1)

axis.scatter(x,y)

axis.set_title('first scatter plot')
axis.set_xlabel('x')
axis.set_ylabel('y')

# これは不要。これをつけることでかえって警告を出してしまう
# fig.show()
# => UserWarning: Matplotlib is currently using module://matplotlib_inline.backend_inline, 
# => which is a non-GUI backend, so cannot show the figure.
```
