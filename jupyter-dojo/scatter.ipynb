## 1. 単純な散布図


```python
import matplotlib.pyplot as plt
import numpy as np

# データを生成する
x = np.random.rand(100)
y = np.random.rand(100)

# 描画領域を作成する
# ひとつの描画領域には複数の座標軸を作成できる
# axes = axisの複数形
figure = plt.figure()

# 座標軸をひとつ作成する
axis = figure.add_subplot(1,1,1)

# 散布図を描く
axis.scatter(x,y)

axis.set_title('first scatter plot')
axis.set_xlabel('x')
axis.set_ylabel('y')

# これは不要。これをつけることでかえって警告を出してしまう
# fig.show()
# => UserWarning: Matplotlib is currently using module://matplotlib_inline.backend_inline, 
# => which is a non-GUI backend, so cannot show the figure.
```

## 2. データごとに色を変える


```python
import matplotlib.pyplot as plt
import numpy as np

1 = np.random.rand(100)*0.5
y1 = np.random.rand(100)

x2 = np.random.rand(100)*0.5 + 0.5
y2 = np.random.rand(100)

figure = plt.figure()

axis = figure.add_subplot(1,1,1)

axis.scatter(x1,y1,c='red')
axis.scatter(x2,y2,c='blue')

axis.set_title('second scatter plot')
axis.set_xlabel('x')
axis.set_ylabel('y')
```

## 3. 凡例とgrid線の追加


```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.random.rand(100)*0.5
y1 = np.random.rand(100)*0.5

x2 = np.random.rand(100)*0.5 + 0.5
y2 = np.random.rand(100)*0.5

x3 = np.random.rand(100)*0.5
y3 = np.random.rand(100)*0.5 + 0.5

x4 = np.random.rand(100)*0.5 + 0.5
y4 = np.random.rand(100)*0.5 + 0.5

figure = plt.figure()

axis = figure.add_subplot(1,1,1)

axis.scatter(x1,y1,c='red', label='group1')
axis.scatter(x2,y2,c='blue', label='group2')
axis.scatter(x3,y4,c='green', label='group3')
axis.scatter(x4,y4,c='yellow', label='group4')



axis.set_title('third scatter plot')
axis.set_xlabel('x')
axis.set_ylabel('y')
axis.grid(True)
axis.legend(loc='upper left')
```
